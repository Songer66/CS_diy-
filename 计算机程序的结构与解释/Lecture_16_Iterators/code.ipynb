```python
# 惰性计算，准备好计算所需的一切，但是只有在被调用的时候再计算
def double(x):
    print('**', x, '=>', 2*x, '**')
    return 2 * x
m = map(double,range(3,7))
list(filter(lambda x : x >=10, m))
```
