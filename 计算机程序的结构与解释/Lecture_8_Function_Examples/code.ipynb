### 8.1 What Would Python Print?

```python
def delay(arg):
    """注意区分返回g函数和直接返回arg的区别"""
    print('delayed')
    def g():
        return arg
    return g # return arg
# delay函数返回了一个返回传入参数的函数，就是等于延迟了一步

# case 1
print('case 1')
delay(delay)()(6)()
# delay(delay)()(6)()
#1.首先delay（delay)
#  先输出delayed，然后返回一个函数g
#  这个函数g的作用时调用时返回一个delay函数
#2.接着就变成了g()(6)()
#  先调用g()，这个时候g(）先打印delayed
#  g()就等价于delay()
#3.接着就变成了delay(6)()
# 先打印delayed
# 然后返回一个函数g，这个函数g的作用是调用是返回6

# case 2
print('case 2')
print(delay(print)()(4))
# 1.首先我们要知道 delay(print)()(4)的值是什么
# 2.delay(print）会先打印delayed
#   然后返回一个函数g，这个函数g就是返回print函数的函数
# 3.接着就变成了g()，就是等价于print
# 4.接着就变成了print(4)，这个时候就会输出4
# 5.最后这个print(4)的值是None，所以最终输出None
```

```python
from operator import add, mul
def square(x):
    return mul(x, x)

def pirate(arggg):
    # 注意这个arggg和plunder函数的参数没有任何关系
    print('matey')
    def plunder(arggg):
        return arggg
    # 我定义了一个函数，这个函数的作用就是返回它的参数
    return plunder #返回这个函数

pirate(3) #这是一个函数,这个函数的作用就是返回它的参数
pirate(3)(square) == pirate(4)(square) == pirate(8)(square)
# case 1
add(pirate(3)(square)(4),1)

# case 2
print('case 2')
pirate(pirate(pirate))(5)(7) == 5(7)
# result，这是一个错误的表达式，因为5这个参数不是一个可执行函数
# 所以对7进行这个函数调用等价于5(7)，这是一个错误的表达式
# 分析
# 拆分，pirate(pirate(pirate))(5)是一个函数
# pirate(pirate(pirate))是一个函数
# pirate(pirate)这个函数就是返回它参数的函数，即一个函数h
# pirate(h)就是一个返回它参数的函数
```

```python
def horse(mask):
    horse = mask # horse这个变量被赋值为mask
    # 这里的mask只是一个函数的名称，horse是一个新的形参
    def mask(horse):
        return horse
    # 这个函数的作用就是返回传入的参数
    return horse(mask) #horse这个函数返回，horse(mask)

mask = lambda horse : horse(2) # 这个mask就是一个lambda函数，用于返回

horse(mask)
```

### 8.2 Implementing Functions

```python
def remove(n,digit):
    """
    Return all digits of non-negative N
    that are not digit, for some
    non-negative Digit less than 10.
    >>> remove(231,3)
    21
    >>> remove(243132,2)
    4313
    """
    kept, digits = 0, 0
    while n > 0:
        n, last = n // 10, n % 10
        if last != digit:
            kept = max(1, kept * 10)
            digits = digits + kept * last
    return digits

remove(243132,2)
```

### 8.3 Decorators 装饰器

```python
# from ucb import trace
import time

def trace1(fn):
    """Return a version of fn that first print it is called"""
    def traced(x):
        print('Calling',fn,'on argument',x)
        return fn(x)
    return traced
# trace1返回的是一个函数，接收的参数是函数fn

def my_eval(fn):
    """这个装饰器用于测量这个函数的执行时间打印日志"""
    def eval_time(x):
        start_time = time.perf_counter()
        result = fn(x)
        end_time = time.perf_counter()
        print(f"Calling {fn} exectued about {end_time-start_time:.5f} second")
        return result
    return eval_time


@trace1
@my_eval
def square(x):
    return x * x
# @trace1
# def square
# square = trace1(square) # 这里会返回一个函数，这个函数会先打印跟踪信息，然后返回square(x)的结果

@trace1
def sum_squares_up_to(n):
    k = 1
    total = 0
    while k <= n:
        total, k = total + square(k), k + 1
    return total


def cache(fn):
    map = {}
    def warp(n):
        if map.get(n) is None:
            map[n] = fn(n)
       
        return map[n]
    return warp



@my_eval
def fibonacii(n):
    return fibonacii_recursive(n)


def fibonacii_iteration(n):
    if n == 0: return 0
    if n == 1: return 1
    current_pos = 1
    current, prev = 1, 0
    while current_pos < n:
        current, prev = current + prev, current
        current_pos += 1
    return current

def fib_linear(n, a=0, b=1):
    """
    n: 还需要计算的步数
    a: 当前数字 (F_i)
    b: 下一个数字 (F_i+1)
    """
    if n == 0:
        return a
    if n == 1:
        return b
    
    # 核心逻辑：n 减 1，同时把 b 传给 a，把 a+b 传给 b
    # 就像在滑动的窗口：[a, b] -> [b, a+b]
    return fib_linear(n - 1, b, a + b)
```
