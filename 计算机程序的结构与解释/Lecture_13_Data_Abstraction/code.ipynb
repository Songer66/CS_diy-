```python
# data abstraction
from math import gcd
def rational(n,d):
    g = gcd(n,d)
    def select(name):
        if name == 'n':
            return n // g
        elif name == 'd':
            return d // g
    return select

def numer(x):
    return x('n')

def denom(x):
    return x('d')

x = rational(3,8)
numer(x)
```

```python
def mul_rational(x,y):
    return rational(numer(x)*numer(y),denom(x)*denom(y))

def add_rational(x,y):
    return rational(numer(x)*denom(y)+numer(y)*denom(x),denom(x)*denom(y))

def sub_rational(x,y):
    return rational(numer(x)*denom(y)-numer(y)*denom(x),denom(x)*denom(y))

def div_rational(x,y):
    n_y = numer(y)
    d_y = denom(y)

    if n_y == 0:
        print("除数不可以为0！")
        return 
    else:
        new_y = rational(d_y,n_y)
        return mul_rational(x,new_y)
    
def print_rational(x):
    return f"{numer(x)} / {denom(x)}"

x = rational(2,5)
y = rational(1,2)
print_rational(mul_rational(x,y))
    
```
