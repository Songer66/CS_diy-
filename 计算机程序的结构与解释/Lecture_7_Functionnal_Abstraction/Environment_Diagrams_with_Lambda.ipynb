```python
def search(f):
    """传入一个函数f,找到一个值满足f(x)为True"""
    x = 0
    while True:
        if f(x):
            return x
        x = x + 1
    # while not f(x):
    #     x = x + 1
    # return x

def is_three(x):
    return x == 3

def square(x):
    return x * x

def positive(x):
    return max(0, square(x) - 200)

# 反函数
def reverse_function(f):
    # 反函数接受一个参数为f的函数，返回一个lambda函数，这个lambda函数接受一个参数
    # 这个参数为y,然后通过search函数找到一个x使得f(x)等于y,最后返回这个x
    return lambda y : search(lambda x : f(x) == y)

inverse = reverse_function(square)
print(inverse(16))
print(search(positive))
print(search(is_three))
```
