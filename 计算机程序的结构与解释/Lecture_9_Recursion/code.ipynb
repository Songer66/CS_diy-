### 9.1 Self-Reference

```python
def print_sums(n):
    print(n)
    def next_sum(k):
        return print_sums(n+k)
    return next_sum 

print_sums(1)(2)(3)
```

### 9.2 Recursive Functions

```python
def split(n):
    return n // 10, n % 10

def sum_digits(n):
    if n < 10: return n 
    all_but_last, last = split(n)
    return sum_digits(all_but_last) + last

# 相互递归的实现
# 138743
# 3 + 4*2 + 7 + (8*2->16->7) + 3 + 2 = 30
def luhn_sum(n):
    if n < 10:
        return n
    else:
        all_but_last, last = split(n)
        return luhn_sum_double(all_but_last) + last

def luhn_sum_double(n):
    all_but_last, last = split(n)
    luhn_digit = sum_digits(2 * last)
    if n < 10:
        return luhn_digit
    else:
        return luhn_sum(all_but_last) + luhn_digit
```
