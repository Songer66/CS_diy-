```python
# list,dict是可变类型对象，tuple是不可变类型对象
def make_withdraw_list(balance):
    b = [balance]
    def withdraw(amount):
        if amount > b[0]:
            return 'Insufficient funds'
        b[0] = b[0] - amount 
        return b[0]
    return withdraw

withdraw = make_withdraw_list(100)
withdraw(50)
a = (1,2)
```
