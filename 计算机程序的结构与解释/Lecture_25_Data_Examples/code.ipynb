```python
s = [2,3]
t = [5,6]
```

```python
s
s.append(t)
s
```

```python
s = [2,3]
t = [5,6]
s.extend(t)
s
```

```python
s = [2,3]
t = [5,6]
a = s + [t]
a
b = a[1:]
b
a[1] = 9
a
b[1][1] = 0
b
```

```python
s = [2,3]
t = [5,6]
s[0:0] = t
s[3:] = t
s
```

```python
# pop removes & returns the last element
# remove removes the first element equal to the argument
# slice assignment can remove elements from a list by assigning [] to a slice
t = [1,2,3]
t[1:3] = [t]
t.extend(t)
t
```

```python
t = [[1,2],[3,4]]
t[0] # 返回的是一个元素
t[1:2] # 返回的是一个新列表
t[0].append(t[1:2])
t
```

```python
# append() 会把你给它的所有内容原封不动地作为一个新元素，添加到列表的末尾。
# extend() 会迭代你给它的对象（必须是可迭代的，比如列表、元组、字符串），把其中的每一个元素逐一取出，依次添加到列表末尾。
```
