```python
# 惰性计算，准备好计算所需的一切，但是只有在被调用的时候再计算
def double(x):
    print('**', x, '=>', 2*x, '**')
    return 2 * x
m = map(double,range(3,7))
list(filter(lambda x : x >=10, m))
```

```python
# 注意不要使用迭代器和list进行比较
t = [1,3,4,3,1]
reversed(t)
reversed(t) == t 
list(reversed(t)) == t
```

```python
# 使用zip
def palindrome(s):
    """
    >>> palindrome('seveneves')
    True
    """
    return all([a == b for a, b in zip(s, reversed(s))])

palindrome('seveneves')
s = 'seveneves'
list(zip(s, reversed(s)))
```
