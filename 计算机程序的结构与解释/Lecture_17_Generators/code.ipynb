```python
def evens(start, end):
    # 如果没有计算的需求，就不会先计算
    even = start + start % 2
    while even < end:
        yield even 
        even += 2
e = evens(1,5)
next(e)
list(e)
```

```python
def a_then_b_1(a,b):
    for x in a:
        yield x
    for x in b:
        yield x

# Generators can Yield from Iterators
def a_then_b_2(a,b):
    yield from a
    yield from b

list(a_then_b_2([3,4],[5,2]))
```

```python
def countdown1(k):
    # 这里返回的是一个生成器，generators
    if k > 0:
        yield k
        yield countdown1(k-1)

def countdown2(k):
    # 正确返回迭代器
    if k > 0:
        yield k
        cd = countdown2(k-1)
        for x in cd:
            yield x

def countdown3(k):
    if k > 0:
        yield k
        yield from countdown3(k-1)
       
cd = countdown3(4)
```

```python
list(cd)
```

```python
# generator是iterator的子集，是一种特殊的迭代器，
# iterator需要手动实现迭代协议，generator通过yield自动实现
# list和dict是可迭代对象，不是迭代器
```

```python
# yield 是生成器函数的核心
# 暂停函数执行，将值返回给调用方
# 下次调用next()的时候，从暂停的位置继续执行

# yield from 可以理解为把一个可迭代对象（
# 生成器、列表、迭代器等的元素逐个yield出来
def prefixes(s):
    if s:
        # yield from会等待子生成器的所有值都yield完
        # 再执行自己的yield s，保证了前缀从短到长的顺序
        yield from prefixes(s[:-1])
        yield s
```

```python
def substrings(s):
    if s:
        yield from prefixes(s)
        yield from substrings(s[1:])
list(substrings('both'))
```
