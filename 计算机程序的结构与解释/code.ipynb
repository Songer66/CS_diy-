### 8.1 What Would Python Print?

```python
def delay(arg):
    """注意区分返回g函数和直接返回arg的区别"""
    print('delayed')
    def g():
        return arg
    return g # return arg


# case 1
print('case 1')
delay(delay)()(6)()
# delay(delay)()(6)()
#1.首先delay（delay)
#  先输出delayed，然后返回一个函数g
#  这个函数g的作用时调用时返回一个delay函数
#2.接着就变成了g()(6)()
#  先调用g()，这个时候g(）先打印delayed
#  g()就等价于delay()
#3.接着就变成了delay(6)()
# 先打印delayed
# 然后返回一个函数g，这个函数g的作用是调用是返回6

# case 2
print('case 2')
print(delay(print)()(4))
# 1.首先我们要知道 delay(print)()(4)的值是什么
# 2.delay(print）会先打印delayed
#   然后返回一个函数g，这个函数g就是返回print函数的函数
# 3.接着就变成了g()，就是等价于print
# 4.接着就变成了print(4)，这个时候就会输出4
# 5.最后这个print(4)的值是None，所以最终输出None
```

```python
from operator import add, mul
def square(x):
    return mul(x, x)

def pirate(arggg):
    # 注意这个arggg和plunder函数的参数没有任何关系
    print('matey')
    def plunder(arggg):
        return arggg
    # 我定义了一个函数，这个函数的作用就是返回它的参数
    return plunder #返回这个函数

pirate(3) #这是一个函数,这个函数的作用就是返回它的参数
pirate(3)(square) == pirate(4)(square) == pirate(8)(square)
# case 1
add(pirate(3)(square)(4),1)

# case 2
print('case 2')
pirate(pirate(pirate))(5)(7) == 5(7)
# result，这是一个错误的表达式，因为5这个参数不是一个可执行函数
# 所以对7进行这个函数调用等价于5(7)，这是一个错误的表达式
# 分析
# 拆分，pirate(pirate(pirate))(5)是一个函数
# pirate(pirate(pirate))是一个函数
# pirate(pirate)这个函数就是返回它参数的函数，即一个函数h
# pirate(h)就是一个返回它参数的函数
```
