```python
def cascade(n):
    if n < 10:
        print(n)
    else:
        print(n)
        cascade(n//10)
        print(n)
```

```python
cascade(112)
```

```python
#1
#12
#123
#12 
#1
def inverse_cascade(n):
    grow(n) # 实现增长打印
    print(n)
    shrink(n) # 实现缩水打印

def f_then_g(f,g,n):
    if n: # 如果n为正值，我们希望先做f函数再做g函数
        f(n) 
        g(n) 

grow = lambda n : f_then_g(grow, print, n//10) # 增长打印，先增长，再打印
shrink = lambda n : f_then_g(print, shrink, n//10) # 缩水打印，先打印，再缩水

inverse_cascade(123456789)
```

### Tree Recursion

```python
# Example:Counting Partitions
# 举例来说我们要计算count_partitions(6,4)
# 需要考虑两种情况，一种情况是用了至少一个4，一种是没有用4
# 即count_partitions(6,4) == count_partitions(2,4) + count_partitions(6,3)
def count_partitions(n,m):
    if n == 0:# 要被分割的数为0，只有一种分法
        return 1
    elif n < 0: # 要被分割的数小于0，出错了，没有这种分法
        return 0
    elif m == 0: # 要使用的分割零件是0，没有这种分法
        return 0
    else:
        with_m_at_least_one = count_partitions(n-m,m)
        without_m = count_partitions(n,m-1)
        return with_m_at_least_one + without_m

count_partitions(3,4)
```
