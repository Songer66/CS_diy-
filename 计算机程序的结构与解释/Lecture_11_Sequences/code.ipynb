### 11.2 For Statements

```python
def count(s,value):
    total = 0
    for element in s:
        if element == value:
            total += 1
    return total

count(digits,1)
```

```python
# Sequence Unpacking in For Statements
pairs = [[1,2],[2,2],[3,2],[4,4]]
same_count = 0
for x,y in pairs:
    if x == y:
        same_count += 1
print(same_count)
```

### 11.3 Ranges 
#### a range is a sequence of consecutive integers

```python
list(range(-2,2)) # 包含起始位置的值，不包含结束位置的值
```

```python
list(range(4)) # 如果就一个参数，则默认从0开始，这个参数的值就是end
```

```python
def sum_below(n):
    total = 0
    for i in range(n):
        total += i
    return total
sum_below(5) # 0,1,2,3,4

def cheer(n):
    for _ in range(n):
        print('Go bears!')

cheer(4)
```

```python
odds = [1,3,5,7,9]
[x+1 for x in odds]
```

```python
[x for x in odds if 25 % x == 0]
```

```python
def divisors(n):
    return [x for x in range(1,n+1) if n % x == 0]
divisors(24)
```
