```python
def min_abs_indices(s):
    """
    >>> min_abs_indices([-4,-3,-2,3,2,4])
    [2,4]
    """
    min_number = 1000000
    results = []
    for i in range(len(s)):
        abs_si = abs(s[i])
        if abs_si == min_number:
            results.append(i)
        elif abs_si > min_number:
            continue
        else:
            results.clear()
            min_number = abs_si
            results.append(i)
    return results

# 简洁版本实现
def min_abs_indices2(s):
    min_abs = min(map(abs,s))
    return [i for i in range(len(s)) if abs(s[i]) == min_abs]

def min_abs_indices3(s):
    min_abs = min(map(abs,s))
    f = lambda i : abs(s[i]) == min_abs
    return list(filter(f,range(len(s))))
print(min_abs_indices3([1,2,3,4,5]))
```

```python
def largest_adj_sum(s):
    """
    >>> largest_adj_sum([-4,-3,-2,3,2,4])
    6
    """
    max_number = -1000000
    for i in range(len(s) - 1):
        j = i + 1
        if s[i] + s[j] >= max_number:
            max_number = s[i] + s[j]
    return max_number

# 简洁实现
def largest_adj_sum(s):
    return max([s[i] + s[i+1] for i in range(len(s) - 1)])

def largest_adj_sum(s):
    return max([a + b for a,b in zip(s[:-1],s[1:])])
s = [-4,-3,-2,3,2,4]
t = [-4,3,-2,-3,2,-4]
largest_adj_sum(t)
```

```python
def digit_dict(s):
    """
    >>> digit_dict([5,8,13,21,34,55,89])
    {1:[21],3:[13], ....}
    """
    res = {}
    for i in range(len(s)):
        t = s[i] % 10 
        if t not in res:
            res[t] = [s[i]]
        else:
            res[t].append(s[i])
    return dict(sorted(res.items(),key = lambda x : x[0]))

# 简洁实现
def digit_dict(s):
    return {d: [x for x in s if x % 10 == d] for d in range(1,10) if d in [x % 10 for x in s]}
digit_dict([5,8,13,21,34,55,89])
```

```python
def all_have_an_equal(s):
    res = {}
    for i in range(len(s)):
        if s[i] not in res:
            res[s[i]] = int(1)
        else:
            res[s[i]] += 1
    return all([x[1] - 1 for x in res.items()])

# 简洁实现
def all_have_an_equal(s):
    return all([s[i] in s[:i] + s[i+1:] for i in range(len(s))])

def all_have_an_equal(s):
    return min([s.count(x) for x in s]) > 1

all_have_an_equal([1,2,3,3,2,1])
```

```python
class Link:
    empty = ()
    def __init__(self, first, rest=empty):
        assert rest is Link.empty or isinstance(rest, Link)
        self.first = first
        self.rest = rest

    def __repr__(self):
        if self.rest:
            rest_repr = ', ' + repr(self.rest)
        else:
            rest_repr = ''
        return 'Link(' + repr(self.first) + rest_repr + ')'

    def __str__(self):
        string = '<'
        while self.rest is not Link.empty:
            string += str(self.first) + ' '
            self = self.rest
        return string + str(self.first) + '>'

def square(x):
    return x * x

def odd(x):
    return x % 2 == 1

def range_link(start, end):
    if start >= end:
        return Link.empty
    else:
        return Link(start, range_link(start + 1,end))

def map_link(f,s):
    if s is Link.empty:
        return s
    else:
        return Link(f(s.first),map_link(f, s.rest))

def filter_link(f,s):
    if s is Link.empty:
        return s
    filtered_rest = filter_link(f, s.rest)
    if f(s.first):
        return Link(s.first, filtered_rest)
    else:
        return filtered_rest

def add(s,v):
    """Add v to an ordered list s with no repeats, returning modified s."""
    assert s is not Link.empty
    if s.first > v:
        s.first, s.rest = v, Link(s.first,s.rest)
    elif s.first < v and s.rest == Link.empty:
        s.rest = Link(v,Link.empty)
    elif s.first < v:
        add(s.rest,v)
    return s

def ordered(s, key=lambda x : x):
    if s is Link.empty or s.rest is Link.empty:
        return True
    elif key(s.first) > key(s.rest.first):
        return False
    else:
        return ordered(s.rest)
    
def merge_in_place(s,t):
    if s is Link.empty:
        return t
    elif t is Link.empty:
        return s
    elif s.first <= t.first:
        # return Link(s.first,merge_in_place(s.rest,t))
        s.rest = merge_in_place(s.rest,t)
        return s
    else:
        # return Link(t.first,merge_in_place(s,t.rest))
        t.rest = merge_in_place(s,t.rest)
        return t
    

```
